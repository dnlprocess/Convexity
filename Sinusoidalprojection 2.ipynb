```python
import numpy as np
import matplotlib.pyplot as plt

def center(X):
    center = np.sum(X, axis=0)/len(X)
    return X - center

def visualize_polygon_wave(polygon, t):
    r_cord = []
    freq = len(polygon)
    r_cord.append([(polygon[i], t[i] * freq * 2 * np.pi) for i in range(len(t))])
    
    x_cord, y_cord = [], []
    for l in range(len(r_cord)):
        x_cord.append([amp * np.cos(theta) for (amp, theta) in r_cord[l]])
        y_cord.append([amp * np.sin(theta) for (amp, theta) in r_cord[l]])
    
    mean_list = []
    
    plt.rcParams["figure.figsize"] = (15, 110)
    for l in range(len(r_cord)):
        plt.plot(int(len(r_cord) / 4) + 1, 4, int(l + 1))
        plt.plot(x_cord[l], y_cord[l])
        plt.plot(np.mean(x_cord[l]), np.mean(y_cord[l]), 'or')
        plt.title("Sampling Freq = " + str(freq) + " Hz")
        
        # Storing the COM for plotting later
        x_mean = np.sum(x_cord[l])
        mean_list.append(x_mean)
    
    plt.show()


triangle = np.array([1 + 1j, 2 + 3j, 3 + 1j])
square = np.array([1 + 1j, 1 + 2j, 2 + 2j, 2 + 1j])
double_square = 1.2 * np.array([1 + 1j, 1 + 2j, 2 + 2j, 2 + 1j, 1 + 1j, 1 + 2j, 2 + 2j, 2 + 1j])
oblong_pentagon = np.array([1 + 1j, 2 + 4j, 4 + 3j, 3 + 1j, 2 + 2j])
concave_shape = np.array([1 + 1j, 2 + 4j, 4 + 3j, 3 + 1j, 2 + 2j, 1 + 2j])
convex_shape = np.array([1 + v + v * 1j for v in np.linspace(0, 2 * np.pi, 10)])

shapes = [triangle, square, oblong_pentagon, double_square, concave_shape, convex_shape]
centered_shapes = [center(shape) for shape in shapes]
for shape in centered_shapes:
    freq = len(shape)
    num_points = len(shape)
    t = np.linspace(0, 2, num=num_points)
    print ("Shape with: " + str(num_points) + " sides")
    visualize_polygon_wave(shape, t)
    visualize_polygon_wave(np.fft.ifft(shape, norm = "ortho"), t)
```
